```python
import gradio as gr
```

```python
# Equation - x3+x2+7x+1
# ax3+bx2+cx+1
```

```python
def equation(x=1, a=1, b=1, c=1, d=1):
    return a*x**3 + b*x**2 + c*x + d
```

```python
interface = gr.Interface(
    fn=equation,
    inputs=["number","number","number","number","number"],
    outputs="number",
    title="Quadratic Polynomial",
    description="Input x,a,b,c,d to output ax^3+bx^2+cx+d"
)
```

```python
interface.launch(share=False, debug=True)
```

```python
# Calculator
```

```python
def calculator(a,b,operation):
    a,b = float(a), float(b)
    if operation == 'Add':
        return a+b
    elif operation == 'Subtract':
        return a-b
    elif operation == 'Multiply':
        return a*b
    elif operation == 'Divide':
        return "Cannot divide by 0" if b==0 else a/b
```

```python
interface = gr.Interface(
    fn=calculator,
    inputs=[
        gr.Number(label="Enter first number"),
        gr.Number(label="Enter second number"),
        gr.Radio(['Add','Subtract','Multiply','Divide'], label='Operation')
    ],
    outputs='text',
    title='Basic Calculator'
)
```

```python
interface.launch(share=False)
```
