## Scientific Calculator
Functions supported -
Add
Subtract
Multiply
Divide
Power
Root
Factorial
Log
Exponential


```python
import gradio as gr
import math
```

```python
def calculator(oper,a,b):
    a, b = float(a), float(b)

    if oper == 'Add':
        return f'{a} + {b} = {a+b}'
    elif oper == 'Subtract':
        return f'{a} - {b} = {a-b}'
    elif oper == 'Multiply':
        return f'{a} * {b} = {a*b}'
    elif oper == 'Divide':
        return 'Cannot divide by 0' if b==0 else f'{a} / {b} = {a/b}'
    elif oper == 'Power':
        return f'{a} raised to power {b} = {a**b}'
    elif oper == 'Root':
        return f'Root {b} of {a} = {a**(1/b)}'
    elif oper == 'Factorial':
        return f'Factorial of {a} = {math.factorial(int(a))}'
    elif oper == 'Natural Logarithm':
        return 'Cannot give Logarithm of 0' if a==0 else f'log({a}) = {math.log(a)}'
    elif oper == 'Custom Logarithm':
        if a==0:
            return 'Cannot give Logarithm of 0'
        elif b==1:
            return 'Cannot give Logarithm for base 1'
        else:
            return f'log({a}) to the base {b} = {math.log(a,b)}'
    elif oper == 'Exponential':
        return f'exp({a}) = {math.exp(a)}'
    else:
        return 'Incorrect Input[s]'
```

```python
interface = gr.Interface(
    fn=calculator,
    inputs=[
        gr.Radio(['Add',
                  'Subtract',
                  'Multiply',
                  'Divide',
                  'Power',
                  'Root',
                  'Factorial',
                  'Natural Logarithm',
                  'Custom Logarithm',
                  'Exponential'],
                 label='Operation'
                 ),
        gr.Number(label='Enter First Number'),
        gr.Number(label='Enter Second Number')
    ],
    outputs='text',
    title='Scientific Calculator',
    description='''
    Input two numbers and operation to compute.
    For Factorial, Natural Logarithm and Exponential, only the first number will be conisdered.
    For Factorial, the first number will be rounded to nearest integer.
    '''
)
```

```python
interface.launch(share=False, debug=True)
```
