```python
# And now - the all important import statement
# If you get an import error - head over to troubleshooting in the Setup folder
# Even for other LLM providers like Gemini, you still use this OpenAI import - see Guide 9 for why

from openai import OpenAI
```

```python
openai = OpenAI()
```

```python
# 1) Identify Business Area
# 2) Identify pain-point
# 3) Propose Agentic AI solution

messages1 = [
    {
        'role':'user',
        'content':'Suggest a business area in the Apparel Retail industry in the United States that might be worth exploring for an Agentic AI opportunity'
    }
]
```

```python
response = openai.chat.completions.create(
    model='gpt-4.1-nano',
    messages=messages1
)

business_area = response.choices[0].message.content
print(business_area)
```

```python
messages2 = [
    {
        'role':'user',
        'content': business_area + ' Identify a pain-point here that would be ripe for an Agentic solution'
    }
]
```

```python
response = openai.chat.completions.create(
    model='gpt-4.1-nano',
    messages=messages2
)

problem_definition = response.choices[0].message.content
print(problem_definition)
```

```python
messages3 = [
    {
        'role':'user',
        'content':problem_definition + ' Propose an Agentic AI solution for the above problem.'
    }
]
```

```python
response = openai.chat.completions.create(
    model='gpt-4.1-nano',
    messages=messages3
)

solution = response.choices[0].message.content
print(solution)
```
